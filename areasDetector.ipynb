```python
import cv2
import numpy as np
import pandas as pd1
import matplotlib.pyplot as plt
import copy
```

```python
img = cv2.imread('rect.png')

class AreaCalculator:
    def __init__(self, img: np.ndarray) -> None:
        self.img_original = img
        self.colored_img = copy.deepcopy(img)
        self.pen_is_down = False

    def draw(self):
        h, w = self.img_original.size
        for j in range(h):
            for i in range(w):
                if self.img_original[i,j] > 200: # if white
                    if self.pen_is_down:
                        self.colored_img[i,j] = 0
                else: # black
                    self.pen_is_down = not self.pen_is_down

        return self.colored_img

a = AreaCalculator(img)
dst = a.draw()

plt.imshow(img)
plt.figure()
plt.imshow(dst)



```

```python
img = cv2.imread('dst.png', cv2.IMREAD_GRAYSCALE) # read the picture

def applay_threshold(img, threshold):
     img[img > threshold] = 255
     img[img <= threshold] = 0
     return img

def fill_the_picture(img):
    pen_is_down = False
    h, w = img.shape
    print(h,w)

    for i in range(h):
        for j in range(w):
            if img[i,j] < 200: # if black
                if  i == w and img[i+1,j] < 200:
                        # black black
                        continue
                else:
                    # black white
                    pen_is_down = not pen_is_down
            
            else: # white
                    if pen_is_down:
                        img[i,j] = 0
            
        pen_is_down = False

    return img

# img = cv2.imread('rect.png', -1)


img = applay_threshold(img, 200)
dst = fill_the_picture(copy.deepcopy(img))

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.figure()
plt.imshow(dst, cmap='gray', vmin=0, vmax=255)
    

# print(img)
```

```python
import random
import cv2
import os
import numpy as np

number_of_images_train = 20
number_of_images_validate = 5
number_of_images_test = 5


image_dimation = (40,40)
width_range = (5, 20)
hight_range = (5, 20)
up_left_point_x_range = \
    (0, image_dimation[0] - width_range[1])
up_left_point_y_range = \
    (0, image_dimation[0] - hight_range[1])

def create_rect_image(base_folder, type: str, name:str):
    input_image = np.ones(image_dimation) * 255
    output_image = np.ones(image_dimation) * 255
    w = random.randint(*width_range)
    h = random.randint(*hight_range)
    up_left_point_x = random.randint(*up_left_point_x_range)
    up_left_point_y = random.randint(*up_left_point_y_range)

    start_point = (up_left_point_x, up_left_point_y)
    end_point = (up_left_point_x + w, up_left_point_y + h)

    input_image = cv2.rectangle(input_image, start_point, end_point, color=(0, 0, 0), thickness=1)
    output_image = cv2.rectangle(output_image, start_point, end_point, color=(0, 0, 0), thickness=-1)

    input_path = os.path.join(base_folder,'A',type,f'{name}.png') 
    output_path = os.path.join(base_folder,'B',type,f'{name}.png') 

    print(f'input path: {input_path}')
    print(f'output path: {output_path}')

    cv2.imwrite(input_path, input_image)
    cv2.imwrite(output_path, output_image)

    return input_image, output_image

root = './datasets/shapes'
os.makedirs(root, exist_ok=True)
os.makedirs(os.path.join(root,'A'), exist_ok=True)
os.makedirs(os.path.join(root,'B'), exist_ok=True)
os.makedirs(os.path.join(root,'A', 'train'), exist_ok=True)
os.makedirs(os.path.join(root,'B', 'train'), exist_ok=True)
os.makedirs(os.path.join(root,'A', 'validate'), exist_ok=True)
os.makedirs(os.path.join(root,'B', 'validate'), exist_ok=True)
os.makedirs(os.path.join(root,'A', 'test'), exist_ok=True)
os.makedirs(os.path.join(root,'B', 'test'), exist_ok=True)

print(f"creating {number_of_images_train} train images")

for i in range(1, number_of_images_train + 1):
    create_rect_image(root,'train', i)

print(f"creating {number_of_images_validate} validate images")
for i in range(1, number_of_images_train + 1):
    create_rect_image(root,'validate', i)

print(f"creating {number_of_images_test} test images")
for i in range(1, number_of_images_train + 1):
    create_rect_image(root,'test', i)

    # input_image = np.ones(image_dimation) * 255
    # output_image = np.ones(image_dimation) * 255
    # w = random.randint(*width_range)
    # h = random.randint(*hight_range)
    # up_left_point_x = random.randint(*up_left_point_x_range)
    # up_left_point_y = random.randint(*up_left_point_y_range)

    # start_point = (up_left_point_x, up_left_point_y)
    # end_point = (up_left_point_x + w, up_left_point_y + h)

    # input_image = cv2.rectangle(input_image, start_point, end_point, color=(0, 0, 0), thickness=1)
    # output_image = cv2.rectangle(output_image, start_point, end_point, color=(0, 0, 0), thickness=-1)
    
    # plt.imshow(input_image, cmap='gray', vmin=0, vmax=255)
    # plt.figure()
    # plt.imshow(output_image, cmap='gray', vmin=0, vmax=255)
    # cv2.imwrite(f'dst{i}.png', ni)    
```
